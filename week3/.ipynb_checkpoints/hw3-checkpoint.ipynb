```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Importing the dataset

```python
dataset = pd.read_csv('D:\mlmgi\data.csv')
X = dataset.iloc[:,1:2].values
#X = X.reshape(-1)
y = dataset.iloc[:,2].values
#y = y.reshape(-1,1)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

# Use Lasso to predict

```python
from sklearn import linear_model
from sklearn.metrics import r2_score

lasso = linear_model.Lasso(alpha=0.1)
lasso.fit(X_train, y_train)
y_pred = lasso.predict(X_test)
Rsq = 'Rsqaure=%.2f' % r2_score(y_test, y_pred)
fig = plt.figure(figsize=(4.0,4.0))
plt.scatter(y_test, y_pred, color = 'gray',label='Data')

Alpha,Beta = np.polyfit(y_test, y_pred,1)
plt.plot(Alpha*y_test+Beta, color='blue')

plt.title('Lasso alpha=0.1')
plt.xlabel('Band Gap [eV]-label')
plt.ylabel('Band Gap [eV]-predict')
plt.xlim(0,8)
plt.ylim(0,8)
plt.text(2.4, 3.6, Rsq)
plt.show()
```

# Use Bayesian to predict

```python
Bayesian = linear_model.BayesianRidge()
Bayesian.fit(X_train, y_train)
y_pred = Bayesian.predict(X_test)
Rsq = 'Rsqaure=%.2f' % r2_score(y_test, y_pred)
fig = plt.figure(figsize=(4.0,4.0))
plt.scatter(y_test, y_pred, color = 'gray',label='Data')

Alpha,Beta = np.polyfit(y_test, y_pred,1)
plt.plot(Alpha*y_test+Beta, color='blue')

plt.title('Bayesian')
plt.xlim(0,8)
plt.ylim(0,8)
plt.xlabel('Band Gap [eV]-label')
plt.ylabel('Band Gap [eV]-predict')
plt.text(2.4, 3.6, Rsq)
plt.show()
```

# Modify Activation

```python
from sklearn.neural_network import MLPRegressor
mlpr_relu =  MLPRegressor(hidden_layer_sizes=(5, 3), activation='relu', solver='adam', alpha=0.0001, batch_size='auto', learning_rate_init=0.001, max_iter=10000, random_state=497)
mlpr_relu.fit(X_train, y_train)
y_pred = mlpr_relu.predict(X_test)
Rsq = 'Rsqaure=%.2f' % r2_score(y_test, y_pred)
fig = plt.figure(figsize=(4.0,4.0))
plt.scatter(y_test, y_pred, color = 'gray',label='Data')
plt.title('MLPRegressor with relu')
plt.xlim(0,8)
plt.ylim(0,8)
plt.xlabel('Band Gap [eV]-label')
plt.ylabel('Band Gap [eV]-predict')
plt.text(2.4, 3.6, Rsq)
plt.show()

mlpr_tanh =  MLPRegressor(hidden_layer_sizes=(5, 3), activation='tanh', solver='adam', alpha=0.0001, batch_size='auto', learning_rate_init=0.001, max_iter=10000, random_state=497)
mlpr_tanh.fit(X_train, y_train)
y_pred = mlpr_tanh.predict(X_test)
Rsq = 'Rsqaure=%.2f' % r2_score(y_test, y_pred)
fig = plt.figure(figsize=(4.0,4.0))
plt.scatter(y_test, y_pred, color = 'gray',label='Data')
plt.title('MLPRegressor with tanh')
plt.xlim(0,8)
plt.ylim(0,8)
plt.xlabel('Band Gap [eV]-label')
plt.ylabel('Band Gap [eV]-predict')
plt.text(2.4, 3.6, Rsq)
plt.show()

mlpr_identity =  MLPRegressor(hidden_layer_sizes=(5, 3), activation='identity', solver='adam', alpha=0.0001, batch_size='auto', learning_rate_init=0.001, max_iter=10000, random_state=497)
mlpr_identity.fit(X_train, y_train)
y_pred = mlpr_identity.predict(X_test)
Rsq = 'Rsqaure=%.2f' % r2_score(y_test, y_pred)
fig = plt.figure(figsize=(4.0,4.0))
plt.scatter(y_test, y_pred, color = 'gray',label='Data')
plt.title('MLPRegressor with identity')
plt.xlim(0,8)
plt.ylim(0,8)
plt.xlabel('Band Gap [eV]-label')
plt.ylabel('Band Gap [eV]-predict')
plt.text(2.4, 3.6, Rsq)
plt.show()

mlpr_logistic =  MLPRegressor(hidden_layer_sizes=(5, 3), activation='logistic', solver='adam', alpha=0.0001, batch_size='auto', learning_rate_init=0.001, max_iter=10000, random_state=497)
mlpr_logistic.fit(X_train, y_train)
y_pred = mlpr_logistic.predict(X_test)
Rsq = 'Rsqaure=%.2f' % r2_score(y_test, y_pred)
fig = plt.figure(figsize=(4.0,4.0))
plt.scatter(y_test, y_pred, color = 'gray',label='Data')
plt.title('MLPRegressor with logistic')
plt.xlim(0,8)
plt.ylim(0,8)
plt.xlabel('Band Gap [eV]-label')
plt.ylabel('Band Gap [eV]-predict')
plt.text(2.4, 3.6, Rsq)
plt.show()
```
