```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Importing the dataset

```python
dataset = pd.read_csv('D:\mlmgi\data.csv')
X = dataset.iloc[:,1:2].values
y = dataset.iloc[:,2].values
```

```python
import math
import matplotlib.gridspec as gridspec

N = len(list(dataset.keys()))
cols = 5
rows = int(math.ceil(N / cols))
#print(N, cols, rows)

fig = plt.figure(figsize=(cols*5, rows*5), facecolor='w')
gs = gridspec.GridSpec(rows, cols)
i=0
feature=dataset[['Density']]
for item in feature.columns:
    ax = fig.add_subplot(gs[i])
    ax.set_title(item, fontsize=18)
    try:
        dataset[item].plot.hist()
    except TypeError:
        print("no numeric data to plot")
    finally:
        i+=1
```

```python
N = len(list(dataset.keys()))
cols = 5
rows = int(math.ceil(N / cols))
#print(N, cols, rows)

fig = plt.figure(figsize=(cols*5, rows*5), facecolor='w')
gs = gridspec.GridSpec(rows, cols)
i=0
target=dataset[['Band Gap']]
for item in target.columns:
    ax = fig.add_subplot(gs[i])
    ax.set_title(item, fontsize=18)
    try:
        dataset[item].plot.hist()
    except TypeError:
        print("no numeric data to plot")
    finally:
        i+=1
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

# Fitting the Regression Model to the dataset

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=100, random_state=0, )
regressor.fit(X_train, y_train)
```

# Visualising the Regression results for higher and smoother curve

```python
X_grid = np.arange(min(X_train), max(X_train), 0.01)
X_grid = X_grid.reshape(len(X_grid), 1)
plt.scatter(X_train, y_train, color = 'gray')
plt.plot(X_grid, regressor.predict(X_grid), color='blue')
plt.title('Random Forest Regression')
plt.xlabel('Density [eV]')
plt.ylabel('Band Gap [eV]')
#density of MoS2 : 5.06 g/cm³
test = np.array([5.06])
test = test.reshape(len(test), 1)
print("The caculated bandgap is: 1.8eV", )
y_pred=regressor.predict(test)
print("The predictive bandgap is: ",y_pred)
plt.scatter(test, y_pred, marker='^', color='red')
plt.show()
```

# Use Lasso to predict

```python
from sklearn import linear_model
from sklearn.metrics import r2_score

lasso = linear_model.Lasso(alpha=0.0041)
lasso.fit(X_train, y_train)
y_pred = lasso.predict(X_test)

lr = linear_model.LinearRegression()
y_test, y_pred=y_test.reshape(len(y_test),1), y_pred.reshape(len(y_pred),1)
lr.fit(y_test, y_pred)
print('Coefficients: \n', lr.coef_)
Rsq = r2_score(y_test, y_pred)
y_test, y_pred=y_test.reshape(len(y_test)), y_pred.reshape(len(y_pred))
fig = plt.figure(figsize=(4.0,4.0))

plt.scatter(y_test, y_pred, color = 'gray',label='Data')
Alpha,Beta = np.polyfit(y_test, y_pred,1)
print(Alpha,Beta)

plt.plot(Alpha*y_test+Beta, color='blue')
#plt.plot(X_grid, lasso.predict(X_grid), color='green',label='pre alpha=0.1')
plt.title('Lasso alpha=0.1')
plt.xlabel('Band Gap [eV]-label')
plt.ylabel('Band Gap [eV]-predict')
plt.xlim(0,8)
plt.ylim(0,8)
plt.text(2.4, 3.6, Rsq)
'''
print("The caculated bandgap is: 1.8eV", )
print("Set alpha=0.1. The predictive bandgap is: ",y_pred)
plt.scatter(test, y_pred, marker='^', color='red',label='pre MoS2,alpha=0.1')
lasso = linear_model.Lasso()
lasso.fit(X_train, y_train)
y_pred = lasso.predict(test)
plt.plot(X_grid, lasso.predict(X_grid), color='blue',label='pre alpha=1')
print("Set alpha=1. The predictive bandgap is: ",y_pred)
plt.scatter(test, y_pred, marker='o', color='red',label='pre MoS2,alpha=1')
plt.legend()
'''
plt.show()
```

```python
Bayesian = linear_model.BayesianRidge()
Bayesian.fit(X_train, y_train)
y_pred = Bayesian.predict(test)
fig = plt.figure(figsize=(8.0,4.0))
plt.scatter(X_train, y_train, color = 'gray',label='Data')
plt.plot(X_grid, Bayesian.predict(X_grid), color='blue',label='pre')
plt.title('Bayesian')
plt.xlabel('Density [eV]')
plt.ylabel('Band Gap [eV]')
print("The caculated bandgap is: 1.8eV", )
print("The predictive bandgap is: ",y_pred)
plt.scatter(test, y_pred, marker='^', color='red',label='pre MoS2')
plt.legend()
plt.show()
```

```python
from sklearn.neural_network import MLPRegressor
mlpr =  MLPRegressor(hidden_layer_sizes=(5, 3), activation='relu', solver='adam', alpha=0.0001, batch_size='auto', learning_rate_init=0.001, max_iter=10000, random_state=497)
mlpr.fit(X_train, y_train)
y_pred = mlpr.predict(test)
fig = plt.figure(figsize=(8.0,4.0))
plt.scatter(X_train, y_train, color = 'gray',label='Data')
plt.plot(X_grid, mlpr.predict(X_grid), color='blue',label='relu')
plt.title('MLPRegressor')
plt.xlabel('Density [eV]')
plt.ylabel('Band Gap [eV]')
print("The caculated bandgap is: 1.8eV", )
print("relu The predictive bandgap is: ",y_pred)
plt.scatter(test, y_pred, marker='^', color='red',label='relu MoS2')

mlpr =  MLPRegressor(hidden_layer_sizes=(5, 3), activation='tanh', solver='adam', alpha=0.0001, batch_size='auto', learning_rate_init=0.001, max_iter=10000, random_state=497)
mlpr.fit(X_train, y_train)
y_pred = mlpr.predict(test)
plt.plot(X_grid, mlpr.predict(X_grid), color='g',label='tanh')
print("tanh The predictive bandgap is: ",y_pred)
plt.scatter(test, y_pred, marker='^', color='red',label='tanh MoS2')

mlpr =  MLPRegressor(hidden_layer_sizes=(5, 3), activation='identity', solver='adam', alpha=0.0001, batch_size='auto', learning_rate_init=0.001, max_iter=10000, random_state=497)
mlpr.fit(X_train, y_train)
y_pred = mlpr.predict(test)
plt.plot(X_grid, mlpr.predict(X_grid), color='black',label='identity')
print("identity The predictive bandgap is: ",y_pred)
plt.scatter(test, y_pred, marker='^', color='red',label='identity MoS2')

mlpr =  MLPRegressor(hidden_layer_sizes=(5, 3), activation='logistic', solver='adam', alpha=0.0001, batch_size='auto', learning_rate_init=0.001, max_iter=10000, random_state=497)
mlpr.fit(X_train, y_train)
y_pred = mlpr.predict(test)
plt.plot(X_grid, mlpr.predict(X_grid), color='y',label='logistic')
print("logistic The predictive bandgap is: ",y_pred)
plt.scatter(test, y_pred, marker='^', color='red',label='logistic MoS2')

plt.legend()
plt.show()
```
